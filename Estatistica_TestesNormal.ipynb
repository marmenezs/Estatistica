Estatística - Testes de Normalidade

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
#criando variável com dados em uma distribuição normal com a função rvs
dados = norm.rvs(size = 1000)
#plotando histograma para verificar o comportamento dos dados
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
#criando qqplot para verificar o comportamento dos dados
figure, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
#executando teste de hipótese (teste de shapiro) para atestar a normalidade dos dados
stats.shapiro(dados)
```

O p é maior que 0,05, ou seja, os dados são normais

Dados não normais

```python
#criando variável com dados em uma distribuição não normais com a função rvs
dados2 = skewnorm.rvs(4, size=1000)
#plotando histograma para verificar o comportamento dos dados
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
#criando qqplot para verificar o comportamento dos dados
figure, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
#executando teste de hipótese (teste de shapiro) para verificar a normalidade dos dados
stats.shapiro(dados2)
```

O p é menor que 0,05, ou seja, os dados não são normais
